### Summary Statistics Based on the Food and Agriculture Organisation of the United Nations.

```python

import pandas as pd

fao_df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv",encoding="latin1")
fao_df

```

```python
#sum of animal fats produced in 2014 and 2017 respectively

animal_fat = fao_df.groupby(['Item'])[['Y2014','Y2017']].sum()
animal_fat
```

```python
# summary statistics of the dataset

fao_stats = fao_df.describe()
fao_stats
```

```python
# total number of missing data in each column

missing_data = fao_df.isnull().sum()
print(missing_data)


# percentage of missing data in each column

percentage = missing_data * 100 /len(fao_df)
percentage
```

```python
# year that had the strongest correlation with 'Element Code'

corr = fao_df.corr()
corr
```

```python
# year that had the highest sum of 'Import Quantity'

max_sum=fao_df.groupby(['Element'])[['Y2014','Y2015','Y2016','Y2017','Y2018']].sum()

max_sum

```

```python
#  sum of 'Production' in 2014

sum_prod = fao_df.groupby(['Element'])[['Y2014']].sum()
sum_prod
```

```python
# 'Element' that had the highest sum and 3rd lowest sum in 2018

emax_sum = fao_df.groupby(['Element'])[['Y2018','Element']].sum().sort_values(by='Y2018',ascending=False)
emax_sum
```

```python
# total 'Import Quantity' in Algeria in 2018

impt_qty = fao_df.groupby(['Area','Element'])[['Y2018']].sum()
impt_qty.head(15)
```

```python
# total number of unique countries in the dataset

u_countries=fao_df.Area.nunique()
u_countries
```
